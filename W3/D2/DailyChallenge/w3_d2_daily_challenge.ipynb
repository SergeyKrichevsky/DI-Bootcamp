```python
image_path = 'Flower Color Images Dataset/flowers/flowers/19_010.png'

from PIL import Image
import matplotlib.pyplot as plt

original_image = Image.open(image_path)
plt.imshow(original_image)
plt.axis('off')
plt.show()
```

Rotate an image by 90 degrees

```python
from scipy.ndimage import rotate
import numpy as np

def rotate_image_30_degrees(image):
    image_array = np.array(image)
    rotated_array = rotate(image_array, 30, reshape=True)
    return rotated_array
```

```python
rotated_30 = rotate_image_30_degrees(original_image)

plt.imshow(rotated_30.astype(np.uint8))
plt.axis('off')
plt.title("Rotated 30")
plt.show()
```

```python
from PIL import ImageOps

mirrored = ImageOps.mirror(original_image)
flipped = ImageOps.flip(original_image)
```

```python
fig, axs = plt.subplots(1, 2, figsize=(10, 5))

axs[0].imshow(mirrored)
axs[0].set_title("Horisontal Flip")
axs[0].axis('off')

axs[1].imshow(flipped)
axs[1].set_title("Vertical Flip")
axs[1].axis('off')

plt.tight_layout()
plt.show()
```

```python
# Step 1: Get original size
width, height = original_image.size

# Step 2: Calculate new size (1.2x zoom)
new_width = int(width * 1.2)
new_height = int(height * 1.2)

# Step 3: Resize (zoom in)
zoomed_image = original_image.resize((new_width, new_height), resample=Image.BICUBIC)

# Display the result
plt.imshow(zoomed_image)
plt.axis('off')
plt.title("Zoomed 1.2x")
plt.show()
```
